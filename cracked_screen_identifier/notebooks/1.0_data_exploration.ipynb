```python

import matplotlib.pyplot as plt
import numpy as np
import glob

# Path to the folder containing your images
image_folder = "cracked_screen_identifier/data/Archive/interim"

# Search for jpg files in the folder
image_filenames = glob.glob(f"{image_folder}/*.jpg")

# Shuffle the list of image filenames
np.random.shuffle(image_filenames)

# Create a figure and a grid of subplots
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(10, 8))

# Flatten the axes array to make it easier to iterate
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < len(image_filenames):
        image = plt.imread(image_filenames[i])
        ax.imshow(image)
        ax.axis('off')
    else:
        ax.axis('off')  # Turn off unused subplots

plt.tight_layout()
plt.show()


```
